#convert 0 and 1 to yes or no

```python
def conv(value):
    if value ==1 :
        return "Yes"
    else:
        return "No"

df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x='SeniorCitizen', data = df, hue='Churn')
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python
# Calculate percentages
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Plot
fig, ax = plt.subplots(figsize=(6, 4))
bottoms = [0] * len(percentages)  # Initialize bottom positions for stacking

# Add bars and annotate
for churn_status, color in zip(percentages.columns, ['#1f77b4', '#ff7f0e']):  # Colors for 'No' and 'Yes'
    bars = ax.bar(
        percentages.index,
        percentages[churn_status],
        bottom=bottoms,
        label=churn_status,
        color=color
    )
    for bar in bars:
        height = bar.get_height()
        if height > 0:  # Annotate only non-zero bars
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{height:.1f}%',
                ha='center',
                va='center',
                fontsize=10
            )
    bottoms += percentages[churn_status]  # Update bottom positions for stacking

# Add labels, title, and legend
ax.set_title("Churn by SeniorCitizen (Stacked Bar Chart with Percentages)")
ax.set_xlabel("Senior Citizen")
ax.set_ylabel("Percentage")
ax.legend(title='Churn', loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=0)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# List of columns to plot
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Number of rows and columns for subplots
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10), constrained_layout=True)

# Plot each column
for i, column in enumerate(columns):
    row, col = divmod(i, n_cols)  # Calculate position in grid
    ax = axes[row, col]
    sns.countplot(x=column, data=df, ax=ax, palette="Set2", hue = df['Churn'])
    ax.set_title(column)
    ax.set_xlabel('')
    ax.set_ylabel('Count')

# Hide unused subplots if any
for j in range(len(columns), n_rows * n_cols):
    fig.delaxes(axes.flat[j])

plt.show()
```

# Customers with Phone Service and Fiber Optic Internet Service show higher churn rates compared to others. Services like Online Security, Online Backup, and Tech Support are associated with lower churn, suggesting these may improve customer retention. Streaming services (Streaming TV and Streaming Movies) show higher churn for users compared to those who don’t use them.
